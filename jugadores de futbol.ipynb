```python
# Actualizamos la columna que creamos de "Nationality_Code"
from iso3166 import countries_by_name
import pytz

# Crear un diccionario manual para los países no encontrados
countries_missing = {
    'ENGLAND': 'GB',
    'WALES': 'GB',
    'BOSNIA HERZEGOVINA': 'BA',
    'KOREA REPUBLIC': 'KR',
    'CZECH REPUBLIC': 'CZ',
    'SCOTLAND': 'GB',
    'CENTRAL AFRICAN REP.': 'CF',
    'DR CONGO': 'CD',
    'IVORY COAST': 'CI',
    'RUSSIA': 'RU',
    'VENEZUELA': 'VE',
    'TURKEY': 'TR',
    'REPUBLIC OF IRELAND': 'IE',
    'UNITED STATES': 'US',
    'CAPE VERDE': 'CV',
    'SYRIA': 'SY',
    'IRAN': 'IR',
    'FYR MACEDONIA': 'MK',
    'CHINA PR': 'CN',
    'GUINEA BISSAU': 'GW',
    'TANZANIA': 'TZ',
    'NORTHERN IRELAND': 'GB',
    'CURACAO': 'CW',
    'MOLDOVA': 'MD',
    'BOLIVIA': 'BO',
    'TRINIDAD & TOBAGO': 'TT',
    'SÃO TOMÉ & PRÍNCIPE': 'ST',
    'KOREA DPR': 'KP',
    'ST KITTS NEVIS': 'KN',
    'ANTIGUA & BARBUDA': 'AG',
    'PALESTINE': 'PS',
    'ST LUCIA': 'LC'
}

# Función para obtener el código del país
def get_country_code(country_name):
    country_name = country_name.upper()
    if country_name in countries_by_name:
        return countries_by_name[country_name].alpha2
    elif country_name in countries_missing:
        return countries_missing[country_name]
    else:
        return 'Unknown'

#  Añadimos la Nueva Columna
df['Nationality_Code'] = df['Nationality'].apply(get_country_code)

# Creamos una nueva funcion que nos permita obtener la zona horaria de cada pais 
def get_time_zone(country_code):
    try:
        return pytz.country_timezones[country_code][0]
    except KeyError:
        return 'Unknown'
    
# Añadir la nueva columna que contenga la zona horaria
df['Time Zone'] = df['Nationality_Code'].apply(lambda x: get_time_zone(x))







```
